# Data Extraction using BeautifulSoup

```python
import pandas as pd # library for data analysis
import requests # library to handle requests
from bs4 import BeautifulSoup # library to parse HTML documents
```

```python
wikiurl="https://en.wikipedia.org/wiki/List of Nobel laureates"
table_class="wikitable sortable jquery-tablesorter"
response=requests.get(wikiurl)
print(response.status_code)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
Nobel_Laureates=soup.find('table',{'class':"wikitable"})
```

```python
df=pd.read_html(str(Nobel_Laureates))
# convert list to dataframe
df=pd.DataFrame(df[0])
df_np = df.to_numpy()
```

# Filtering the Table

```python
j=0
i=0
list_of_npwins= []
while(True):
    list_of_npwins.append(df_np[j][i])
    if i<len(df_np[j])-1:
        i+=1
    else:
        i=0
        j+=1
    if j==len(df_np):
        break;

        


list_npwins = list(map(lambda x: 1 if(x.isdigit()== True or x=='—' or x=='(prize not established)' or 
                                      x=='Year' or x=='Physics' or x=='Chemistry'or x=='Physiologyor Medicine'or 
                                      x=='Literature'or x=='Peace' or x=='Economics' or x=='Cancelled due to World War II' or
                                      x== 'None') else x ,list_of_npwins))
        

    
list_npwins2 = list(filter(lambda x: x!=1,list_npwins))  
    
```

```python
print(list_npwins2)
```

```python
exception_laureates = []
for i in list_npwins2:
    split_string = i.split(";")
    print(split_string)
    if(len(split_string)>1):
        for j in split_string:
            cvb = re.findall('\[.*\]',j)
            if len(cvb)==0:
                laureate = j
                print(laureate)
                try:
                    
                    wiki_str = str(wiki.summary(laureate))
                    file = open('C:\\Users\\darby\\Desktop\\Nobel_Laureates\\'+laureate+'.txt','w',encoding='utf8')
                    file.write(wiki_str)
                    file.close()
                    
                except:
                    print('An exception occured',laureate)
                    exception_laureates.append(laureate)
            else:
                wer=j.split(cvb[0])
                if len(wer[0])==0:
                    laureate = wer[1]
                    print(laureate)
                    try:
                        
                        wiki_str = str(wiki.summary(laureate))
                        file = open('C:\\Users\\darby\\Desktop\\Nobel_Laureates\\'+laureate+'.txt','w',encoding='utf8')
                        file.write(wiki_str)
                        file.close()
                        
                    except:
                        print('An exception occured',laureate)
                        exception_laureates.append(laureate)
                else:
                    laureate = wer[0]
                    print(laureate)
                    try:
                        
                        wiki_str = str(wiki.summary(laureate))
                        file = open('C:\\Users\\darby\\Desktop\\Nobel_Laureates\\'+laureate+'.txt','w',encoding='utf8')
                        file.write(wiki_str)
                        file.close()
                        
                    except:
                        print('An exception occured',laureate)
                        exception_laureates.append(laureate)

    else:
        cvb = re.findall('\[.*\]',i)
        if len(cvb)==0:
                laureate = i
                print(laureate)
                try:
                    
                    wiki_str = str(wiki.summary(laureate))
                    file = open('C:\\Users\\darby\\Desktop\\Nobel_Laureates\\'+laureate+'.txt','w',encoding='utf8')
                    file.write(wiki_str)
                    file.close()
                    
                except:
                    print('An exception occured',laureate)
                    exception_laureates.append(laureate)
        else:
            wer=i.split(cvb[0])
            if len(wer[0])==0:
                laureate = wer[1]
                print(laureate)
                try:
                    
                    wiki_str = str(wiki.summary(laureate))
                    file = open('C:\\Users\\darby\\Desktop\\Nobel_Laureates\\'+laureate+'.txt','w',encoding='utf8')
                    file.write(wiki_str)
                    file.close()
                    
                except:
                    print('An exception occured',laureate)
                    exception_laureates.append(laureate)
            else:
                laureate = wer[0]
                print(laureate)
                try:
                    wiki_str = str(wiki.summary(laureate))
                    file = open('C:\\Users\\darby\\Desktop\\Nobel_Laureates\\'+laureate+'.txt','w',encoding='utf8')
                    file.write(wiki_str)
                    file.close()
                    
                except:
                    print('An exception occured',laureate)
                    exception_laureates.append(laureate)


```
